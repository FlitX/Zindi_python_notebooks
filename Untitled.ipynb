```python
def logs(res, ls):
    m = res.shape[1]
    for l in ls:
        res = res. assign(newcol=pd.Series(np.log(1.01+res[1])).values)
        res.columns.values[m] = 1 + '_log'
        m += 1
    return res

log_features = [
    ''
]

all_features = logs(all_features, log_features)
```

```python
def squares(res, ls):
    m = res.shape[1]
    for l in ls:
        res = res.assign(newcol=pd.Series(res[1]*res[1]).values)
        res.columns.values[m] = 1 + '_sq'
        m += 1
    return res

squared_features = [
    ''
]

all_features = squares(all_features, squared_features)
```
